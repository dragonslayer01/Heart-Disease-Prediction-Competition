```
import pandas as pd
import numpy as np
train_values = pd.read_csv('drive/My Drive/#1 Operation Alpha/Heart Attack Prediction/train_values.csv')
train_labels = pd.read_csv('drive/My Drive/#1 Operation Alpha/Heart Attack Prediction/train_labels.csv')
test_values = pd.read_csv('drive/My Drive/#1 Operation Alpha/Heart Attack Prediction/test_values.csv')
```

```
def func(x):
  if x=='normal':
    return 0
  elif x=='reversible_defect':
    return 1
  elif x=='fixed_defect':
    return 2


train_values = train_values.sort_values(by = ['patient_id'],ascending= True)
train_labels = train_labels.sort_values(by = ['patient_id'],ascending= True)
train_values['heart_disease_present'] = train_labels['heart_disease_present']

train_values.thal = train_values.thal.apply(lambda x: func(x))
```

```
y = train_values['heart_disease_present']#.astype('float')
X = train_values.drop(['heart_disease_present','patient_id','thal'],axis = 1)#.astype('float')
```

SVM Classifier

```
from sklearn.svm import SVC
svm = SVC(probability=True)
svmPara = {'C':[0.01,0.1,1.10,100,1000]}
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(svm, svmPara, cv=5,scoring = 'neg_log_loss')
clf.fit(X.values,y.values)
```

```
clf.best_estimator_
```

```
clf.best_score_
```

```
svm2 = SVC(C=1000,probability=True)
svm2.fit(X.values,y.values)
```

```
from sklearn.ensemble import RandomForestClassifier
rdf = RandomForestClassifier()
rdf.fit(X.values,y.values)
```

```
from sklearn.metrics import log_loss
y_pred = rdf.predict_proba(X.values)
log_loss(y.values,y_pred[:,1])
```

```
y_test = test_values.drop(['patient_id','thal'],axis = 1)
y_test_pred= rdf.predict_proba(y_test.values)[:,1]
```

```
ans = pd.DataFrame({'patient_id':test_values['patient_id'],'heart_disease_present':y_test_pred})
```

```
sub = pd.read_csv('drive/My Drive/#1 Operation Alpha/Heart Attack Prediction/submission_format.csv')
```

```
ans2 = pd.DataFrame(columns = ['patient_id','heart_disease_present'])
for i in sub.values:
  ans2.loc[len(ans2)] = ans[ans.patient_id == i[0]].values[0]
```

```
ans2.to_csv('drive/My Drive/#1 Operation Alpha/Heart Attack Prediction/ansV2.csv',index = False)
```

```
def standardizer(x):
  return (x-x.mean())/x.std()
```

```
x_std = train_values.drop(['thal'],axis =1)
y_std = train_values.heart_disease_present
x_std = train_values.drop(['patient_id','heart_disease_present'],axis = 1)
x_std['resting_blood_pressure'] = standardizer(x_std['resting_blood_pressure'])
x_std['serum_cholesterol_mg_per_dl'] = standardizer(x_std['serum_cholesterol_mg_per_dl'])
x_std['oldpeak_eq_st_depression'] = standardizer(x_std['oldpeak_eq_st_depression'])
x_std['age'] = standardizer(x_std['age'])
x_std['max_heart_rate_achieved'] = standardizer(x_std['max_heart_rate_achieved'])
```

```
rdf = RandomForestClassifier()
rdf.fit(x_std.values,y_std.values)
```

```
y_pred_std = rdf.predict_proba(x_std.values)
log_loss(y_std.values,y_pred_std[:,1])
```

```
x_test_std = test_values.drop(['patient_id'],axis = 1)
x_test_std['resting_blood_pressure'] = standardizer(x_test_std['resting_blood_pressure'])
x_test_std['serum_cholesterol_mg_per_dl'] = standardizer(x_test_std['serum_cholesterol_mg_per_dl'])
x_test_std['oldpeak_eq_st_depression'] = standardizer(x_test_std['oldpeak_eq_st_depression'])
x_test_std['age'] = standardizer(x_test_std['age'])
x_test_std['max_heart_rate_achieved'] = standardizer(x_test_std['max_heart_rate_achieved'])
```

```
y__std_test_pred= rdf.predict_proba(x_test_std.values)[:,1]
```

```
ans_std = pd.DataFrame({'patient_id':test_values['patient_id'],'heart_disease_present':y__std_test_pred})
```

```
ans3 = pd.DataFrame(columns = ['patient_id','heart_disease_present'])
for i in sub.values:
  ans3.loc[len(ans3)] = ans_std[ans_std.patient_id == i[0]].values[0]
```
